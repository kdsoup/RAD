```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Task 7

estimates = pd.read_csv('task7_estimates.csv', header=None).values.flatten()
sorted_estimates = sorted(estimates)
exact_S = 6980

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(sorted_estimates) + 1), sorted_estimates, marker='o', linestyle='-', label='Sorted Estimates')
plt.axhline(y=exact_S, color='r', linestyle='--', label='Exact S')
plt.xlabel('Experiment Index (sorted)')
plt.ylabel('Estimate Value')
plt.title('Task 7: Sorted Estimates vs Exact S')
plt.legend()
plt.grid(True)
plt.show()

medians = pd.read_csv('task7_medians.csv', header=None).values.flatten()
sorted_medians = sorted(medians)

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(sorted_medians) + 1), sorted_medians, marker='o', linestyle='-', label='Sorted Medians')
plt.axhline(y=exact_S, color='r', linestyle='--', label='Exact S')
plt.xlabel('Group Index (sorted)')
plt.ylabel('Median Value')
plt.title('Task 7: Sorted Medians vs Exact S')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Task 8

task8_df = pd.read_csv('task8_results.csv')

plt.figure(figsize=(10, 6))
plt.plot(task8_df['m'], task8_df['mse'], marker='o', linestyle='-', label='Mean Square Error')
plt.xlabel('Number of Counters (m)')
plt.ylabel('Mean Square Error')
plt.title('Task 8: MSE vs m')
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(task8_df['m'], task8_df['runtime'], marker='o', linestyle='-', label='Runtime (ms)', color='orange')
plt.xlabel('Number of Counters (m)')
plt.ylabel('Runtime (ms)')
plt.title('Task 8: Runtime vs m')
plt.grid(True)
plt.legend()
plt.show()
```
